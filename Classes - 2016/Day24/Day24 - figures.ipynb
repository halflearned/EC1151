```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("white")
```

```python
muhat = 1
mu_hyp = 0
sigmasq = 2
sigma = np.sqrt(2)
n = 10
alpha = 0.05
```

```python
(muhat - mu_hyp)/np.sqrt((sigmasq)/n)
```

```python
x = np.linspace(-3.5, 3.5, 1000)
a = np.linspace(0.01, .99, 1000)

Finv = ss.norm(mu_hyp, sigma).ppf(a)
F = ss.norm(mu_hyp, sigma).cdf(x)

m1 = ss.norm(mu_hyp, sigma).ppf(alpha/2)
m2 = ss.norm(mu_hyp, sigma).ppf(1 - alpha/2)

y1 = ss.norm(mu_hyp, sigma).cdf(m1)
y2 = ss.norm(mu_hyp, sigma).cdf(m2)


fig, ax = plt.subplots(1, figsize = (8, 6))
ax.plot(x, F)

ax.set_yticks([.05, .25, .5, .75, .95])
ax.set_yticklabels(["$\\alpha/2$", .25, .5, .75, "$1 - \\alpha/2$"], fontsize = 18)
ax.set_xticks([m1, 0, m2])
ax.set_xticklabels(["$m_1$", 0, "$m_2$"], fontsize = 20)

lxlim = ax.get_xlim()[0]

ax.vlines(x = m1, ymin = 0, ymax = y1, linestyle = "--")
ax.vlines(x = m2, ymin = 0, ymax = y2, linestyle = "--")
ax.hlines(y = y1, xmin = lxlim, xmax = m1, linestyle = "--")
ax.hlines(y = y2, xmin = lxlim, xmax = m2, linestyle = "--")
ax.set_ylim((0,1.1))

ax.set_title("CDF of $\hat{\mu}$", fontsize = 20)


sns.despine()

fig.savefig("cdf")
```

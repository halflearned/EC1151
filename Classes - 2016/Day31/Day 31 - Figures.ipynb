```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
import pandas as pd
sns.set_style("white")
%matplotlib inline
```

```python
x = np.random.randint(1,7, size = 150)
y = np.random.uniform(1,3, size = 150)
```

```python
n = 1000
ygrid = np.linspace(0, 10, n)
ypdf = []
ypdf += [ss.norm(5, 1).pdf(ygrid)]
ypdf += [ss.norm(5, 2).pdf(ygrid)]
ypdf += [ss.norm(5, .1).pdf(ygrid)]
ypdf += [ss.uniform(2, 6).pdf(ygrid)]
colors = np.random.uniform(size= (len(ypdf), 3))
```

```python
for k, ys in enumerate(ypdf):

    fig, ax = plt.subplots(1, 2, figsize = (15, 4))
    ax[0].plot(ygrid, ypdf[0], linewidth = 5)
    ax[0].set_xticks([5])
    ax[0].set_xticklabels([5], fontsize = 20)
    ax[0].set_title("$P(Y|X = x)$", fontsize = 30)

    ax[1].plot(ygrid, ys, color = colors[k], linewidth = 5)
    ax[1].set_xticks([5])
    ax[1].set_xticklabels([5], fontsize = 20)
    ax[1].set_title("$P(Y|X = \\tilde{x})$", fontsize = 30)
    sns.despine()

    fig.savefig("norm_xy_indep_%d" % k)
```

```python
fig, axs = plt.subplots(2, 2, figsize = (20, 15))
for ys, ax,c in zip(ypdf, axs.flatten(), colors):
    ax.plot(ygrid, ys, linewidth = 5, color = c)
    ax.vlines(5, *ax.get_ylim(), linestyle = "--", color = "red", linewidth = 3)
    ax.set_xticks([5])
    ax.set_xticklabels([5], fontsize = 20)

sns.despine()
fig.savefig("norm_xy_indep")
```

$$E[Y] = 
\text{Avg of} 
\begin{cases} 
    E[Y|X=60] = \sum_i y_i \cdot p_{Y|X}(y_i \ | \ 60) \\
    E[Y|X=70] = \sum_i y_i \cdot p_{Y|X}(y_i \ | \ 70) 
\end{cases}$$


$$E[Y] = p_X(60)E[Y|X=60] + p_X(70)E[Y|X=70]$$

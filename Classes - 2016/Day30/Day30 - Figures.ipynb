```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
%matplotlib inline
```

```python
X = np.array([1,2,3,4,5]*10)
Y = ss.norm(loc = X).rvs()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].scatter([0]*50, Y, s = 200, alpha = 0.4)
ax[0].set_xticklabels("")
ax[0].set_title("Realizations of $Y$, without any information", fontsize = 20)
fig.savefig("scatter1")
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].scatter([0]*50, Y, s = 200, alpha = 0.4)
ax[0].set_xticklabels("")
ax[0].scatter(0, np.mean(Y), s = 700, marker = "*", color = "red")
ax[0].set_title("Realizations of $Y$, without any information", fontsize = 20)
fig.savefig("scatter2")
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].scatter([0]*50, Y, s = 200, alpha = 0.4)
ax[0].set_xticklabels("")
ax[0].scatter(0, np.mean(Y), s = 700, marker = "*", color = "red")
ax[0].set_title("Realizations of $Y$, without any information", fontsize = 20)
ax[1].scatter(X, Y, s = 200, alpha = 0.4)
ax[1].set_title("Realizations of $Y$, knowing $X$", fontsize = 20)
ax[1].set_ylim(-2, 8)
ax[1].set_xticks([1,2,3,4,5])
ax[1].set_xticklabels([1,2,3,4,5], fontsize = 20);
fig.savefig("scatter3")
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].scatter([0]*50, Y, s = 200, alpha = 0.4)
ax[0].scatter(0, np.mean(Y), s = 500, marker = "*", color = "red", label = "Best guess: mean")
ax[0].set_xticklabels("")
ax[0].set_title("Realizations of $Y$, without any information", fontsize = 20)
ax[1].scatter(X, Y, s = 200, alpha = 0.4)
for x in [1,2,3,4,5]:
    ax[1].scatter(x, np.mean(Y[X == x]), s = 700, marker = "*", color = "red")
ax[1].set_title("Realizations of $Y$, knowing $X$", fontsize = 20)
ax[1].set_ylim(-2, 8)
ax[1].set_xticks([1,2,3,4,5])
ax[1].set_xticklabels([1,2,3,4,5], fontsize = 20);
fig.savefig("scatter4")
```

```python
Y = [100, 80, 60]
X = [1, 3, 10]
```

```python
A = np.array([[.05, .05, .2], [.05, .1, .1], [.2, .15 ,.1]])
px = np.sum(A, 0)
py = np.sum(A, 1)
EX = np.sum(px*X)
EY = np.sum(py*Y)
```

```python
cov = 0
for i in range(3):
    for j in range(3):
        print("(",X[i],"-",EX,")(",Y[j],"-",EY,") = ", (X[i] - EX)*(Y[j] - EY),"\qquad wp \qquad p(",
              X[i],",",Y[j],") = ", A[j,i], "\\\\")
        cov += (X[i] - EX)*(Y[j] - EY)*A[j,i]
```

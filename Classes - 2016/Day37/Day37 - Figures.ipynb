```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 1000)
u = ss.uniform(-1.75, 3.5)
g = ss.gamma(1, loc = -1, scale = 1)
t = ss.laplace(loc = 0, scale = np.sqrt(.5))
n = ss.norm(0, 1)
```

```python
fig, ax = plt.subplots(1, figsize = (9, 6))
for v in [u, g, n, t]:
    ax.plot(x, v.pdf(x), linewidth = 4, alpha = .7)
sns.despine()
ax.set_xticklabels("")
ax.set_yticklabels("")
fig.savefig("dists1")
```

```python
fig, ax = plt.subplots(1, figsize = (9, 6))
for v in [n]:
    ax.plot(x, v.pdf(x), linewidth = 4, alpha = .7)
sns.despine()
ax.set_xticklabels("")
ax.set_yticklabels("")
fig.savefig("dists2")
```

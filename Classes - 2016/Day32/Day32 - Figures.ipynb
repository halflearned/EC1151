```python
import numpy as np
import pandas as pd
import scipy.stats as ss
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set_style("white")
%matplotlib inline
```

```python
xgrid = np.linspace(-5, 5, 100)
f = np.sin(2*xgrid)
g = 2*xgrid
```

```python
fig, ax = plt.subplots(1,2, figsize = (15, 4), sharey = True, sharex = True)
ax[0].plot(xgrid, f, linewidth = 2)
ax[0].set_title("$f$", fontsize = 25)
ax[0].set_xticklabels("")
ax[0].set_yticklabels("")
ax[1].plot(xgrid, 3*f, linewidth = 2, color = "darkgreen")
ax[1].set_title("$3f$", fontsize = 25)
ax[1].set_xticklabels("")
ax[1].set_yticklabels("")
sns.despine()
fig.savefig("funcscalarmult")
```

```python
fig, ax = plt.subplots(1,3, figsize = (15, 4), sharey = True)
ax[0].plot(xgrid, f, linewidth = 2)
ax[0].set_title("$f$", fontsize = 25)
ax[0].set_xticklabels("")
ax[0].set_yticklabels("")
ax[1].plot(xgrid, g, linewidth = 2, color = "darkgreen")
ax[1].set_title("$g$", fontsize = 25)
ax[1].set_xticklabels("")
ax[1].set_yticklabels("")
ax[2].plot(xgrid, f + g, linewidth = 2, color = "darkred")
ax[2].set_title("$f + g$", fontsize = 25)
ax[2].set_xticklabels("")
ax[2].set_yticklabels("")
sns.despine()
fig.savefig("funcsum")
```

```python
X = np.random.uniform(0,10, size = 50)
Y = 2 + 3*X + np.random.normal(0, 6, size = 50)
xgrid= np.linspace(X.min(), X.max(), 100)
```

```python
fig, ax = plt.subplots(1)
ax.scatter(X, Y, s = 200, alpha = .5)
ax.scatter(4, 8, s = 200, alpha = .5, color = "blue")
ax.scatter(7, 28, s = 200, alpha = .5, color = "blue")
ax.set_xlabel("X", fontsize = 20)
ax.set_ylabel("Y", fontsize = 20, rotation = 0)
ax.set_xticklabels("")
ax.set_yticklabels("")
sns.despine()
fig.savefig("decomposition1")
```

```python
fig, ax = plt.subplots(1)
ax.scatter(X, Y, s = 200, alpha = .1)
ax.scatter(4, 8, s = 200, alpha = .5, color = "red")
ax.scatter(7, 28, s = 200, alpha = .5, color = "red")
ax.plot([4, 4], [8, 2 + 3*4])
ax.plot([7, 7], [28, 2 + 3*7])
ax.plot(xgrid, 2 + 3*xgrid, linestyle = "-", linewidth = 4, color = "darkorange")
ax.set_xlabel("X", fontsize = 20)
ax.set_ylabel("Y", fontsize = 20, rotation = 0)
ax.set_xticklabels("")
ax.set_yticklabels("")
sns.despine()
fig.savefig("decomposition2")
```

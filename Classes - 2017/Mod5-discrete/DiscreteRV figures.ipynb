Bernoulli

```python
r = []
for z in range(0, 7):
    r += [np.linspace(z+.15,z+.85, 100)]
r = np.hstack(r)
```

```python
p = .1
fig, ax = plt.subplots(1)
x = [1,2,3,4,5,6]
y = [1/6]*len(x)
ax.scatter(x, y, s= 100)
x1 = np.linspace(0, 7, 100)
ax.stem(r, [0]*len(r), linefmt='b-', markerfmt='bo', basefmt='o')
ax.scatter(x, [0]*6, s= 100, c= "white")

#ax.stem(x, y, linefmt='', markerfmt='bo', basefmt='')
ax.set_xlim([0, 7]);
ax.set_ylim([0, .2]);
ax.set_xticks(x);
ax.set_xticklabels(x, fontsize = 20);
ax.set_yticks([1/len(x)]);
ax.set_yticklabels(["1/6"], fontsize= 20);
ax.set_title("DiscreteUniform[1,6]", fontsize = 20);
sns.despine()
fig.savefig("discrete1")
```

```python
fig, ax = plt.subplots(1)
a = 1
b = 3
x = range(a, b+1)
y = [1/len(x)]*len(x)
#ax.scatter(x, y, s= 100)
ax.stem(x, y, linefmt='b-', markerfmt='bo', basefmt='o')
ax.set_xlim([a-1, b+1]);
ax.set_ylim([0, y[0] + .1]);
ax.set_xticks(x);
ax.set_xticklabels(x, fontsize = 20);
ax.set_yticks([1/len(x)]);
ax.set_yticklabels([str(np.round(y[0],3))]*len(x), fontsize= 20);
ax.set_title("DiscreteUniform[%d,%d]" % (a,b), fontsize = 20);
sns.despine()
fig.savefig("discrete3")
```

```python
p = .1
fig, ax = plt.subplots(1)
x = [0,1]
y = [1-p, p]
ax.scatter(x, y, s= 100)
#ax.stem(x, y, linefmt='', markerfmt='bo', basefmt='')
ax.set_xlim([-.5,1.5]);
ax.set_ylim([0, 1]);
ax.set_xticklabels(["",0,"",1,""]);
ax.set_yticks(np.linspace(0,1,11));
ax.grid()
ax.set_title("$Bern(%2.1f)$" % p, fontsize = 20);
```

```python
p = .1
fig, ax = plt.subplots(1)
x = [0,1]
y = [1-p, p]
ax.stem(y, linefmt='b-', markerfmt='bo', basefmt='r-')
ax.stem(x, y, linefmt='b-', markerfmt='bo', basefmt='r-')
ax.set_xlim([-.5,1.5]);
ax.set_ylim([0, 1]);
ax.set_xticklabels(["",0,"",1,""]);
ax.set_yticks(np.linspace(0,1,11));
ax.grid()
ax.set_title("$Bern(%2.1f)$" % p, fontsize = 20);
```

Binomial

```python
from scipy.misc import comb
```

```python
p = 0.5
n = 5
fig, ax = plt.subplots(1)
x = range(n+1)
#y = [comb(n, k)*(p**k)*(1-p)**(n-k) for k in range(0, n+1)]
y = [ss.binom(n, p).pmf(z) for z in x]
ax.stem(x, y, linefmt='b', markerfmt='bo')
ax.set_title("$Bin(%d, \ %2.3f)$" % (n, p), fontsize = 20);
sns.despine()
fig.savefig("bin4.png")
```
